```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../Dataset/Training Data.csv")

print(df.info())
print(df["Risk_Flag"].value_counts())
```

```python
import matplotlib.pyplot as plt


# Risk_Flag 
risk_counts = df["Risk_Flag"].value_counts()
labels = ["No Risk (0)", "Risk (1)"]
colors = ["#66b3ff", "#ff9999"]  # 
explode = (0.05, 0.05)

# 
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count:,})'
    return my_autopct

# 
plt.figure(figsize=(10, 8))
patches, texts, autotexts = plt.pie(risk_counts, 
                                  labels=labels, 
                                  autopct=make_autopct(risk_counts),
                                  startangle=140, 
                                  colors=colors, 
                                  explode=explode,
                                  shadow=False,
                                  wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})

# 
# plt.title("Loan Risk Distribution Analysis", fontsize=16, fontweight='bold', pad=20)
plt.suptitle("Distribution of Risk Flags in Loan Dataset", fontsize=12, y=0.95)

# 
plt.legend(patches, labels, 
          title="Risk Categories",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# 
plt.figtext(0.99, 0.01, "Data Source: Loan Dataset", 
           ha='right', va='bottom', 
           fontsize=8, style='italic')

plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python

import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("../Dataset/Training Data.csv")

numeric_df = df.drop(columns=['Id']).select_dtypes(include=['int64', 'float64'])

# 
corr_matrix = numeric_df.corr()

# 
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", square=True, cbar_kws={"shrink": 0.8})
plt.title("Risk Flag Correlation Matrix", fontsize=14)
plt.tight_layout()
plt.show()
```
